# Modeling 

```python
#I want to focus on what we can control to provide more tangible suggestions to Seattle
#I'll start with some k means clustering to see if there are specific areas with collisions and from there we will explore severity
#run k means on non severe 
#run k means on severe accidents too 
#observe overlap
#remove observations without coordinates
k_collisions=collisions.copy(deep=True)
k_collisions.dropna(axis=0,how='any',subset=['X','Y'],inplace=True)
```

```python
# Variable with the Longitude and Latitude and objectID
X=k_collisions.loc[:,['OBJECTID','X','Y']]
X.head(10)
```

```python
K_clusters = range(1,10)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
Y_axis = k_collisions[['X']]
X_axis = k_collisions[['Y']]
score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]
# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
#we observe a drop off after 4 clusters so I chose k=4 
kmeans = KMeans(n_clusters = 4, init ='k-means++')
kmeans.fit(X[X.columns[1:4]]) # Compute k-means clustering.
X['cluster_label'] = kmeans.fit_predict(X[X.columns[1:4]])
centers = kmeans.cluster_centers_ # Coordinates of cluster centers.
labels = kmeans.predict(X[X.columns[1:4]]) # Labels of each point
X.head(10)
```

```python
#graph the resulting clusters
X.plot.scatter(x = 'X', y = 'Y', c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
#now lets split the set into severe and non-severe
severe_collisions= k_collisions[k_collisions['SEVERITYCODE'] == 1]

K_clusters1 = range(1,10)
kmeans1 = [KMeans(n_clusters=i) for i in K_clusters1]
Y_axis1 = severe_collisions[['X']]
X_axis1 = severe_collisions[['Y']]
score1 = [kmeans1[i].fit(Y_axis1).score(Y_axis1) for i in range(len(kmeans1))]
# Visualize
plt.plot(K_clusters1, score1)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```
